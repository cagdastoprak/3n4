```
import numpy as np
import matplotlib.pyplot as plt
```

```
mean1=[]
for i in range(50):
#Birinci
  birinci=np.random.normal(100,15,1000)
  mean1.append(birinci.mean())
print(mean1)
print('len :',len(mean1))
```

```
mean2=[]
for i in range(1000):
#İkinci
  ikinci=np.random.normal(100,15,10000)
  mean2.append(ikinci.mean())
print(mean2)
print('len :',len(mean2))
```

```
plt.hist(mean1, bins = 25)
```

```
plt.hist(mean2,bins = 25)
```

```
adilZar = np.random.binomial(1, .5, 1000)
plt.hist(adilZar)
print('mean2 =',adilZar.mean())
```

```
hileliZar = np.random.binomial(1, .75, 1000)
plt.hist(hileliZar)
print('mean2 =',hileliZar.mean())
```

```
hileliZar = np.random.binomial(1, .75, 1000000) # Fark git gide açılıyor, o yüzden deneme sayısı arttıkça hangi zar olduğunu bilmemiz kolaylaşıyor
plt.hist(hileliZar)
print('mean2 =',hileliZar.mean())
```
